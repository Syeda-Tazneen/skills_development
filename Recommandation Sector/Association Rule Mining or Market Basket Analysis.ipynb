```python
dataset = pd.read_csv("data-2.csv")
dataset.head()
```

```python
# Data Cleaning Part
dataset['Description'] = dataset['Description'].str.strip()

# removes duplicate invoice
dataset.dropna(axis=0, subset=['InvoiceNo'], inplace=True)

# converting invoice number to be string
dataset['InvoiceNo'] = dataset['InvoiceNo'].astype('str')
# remove the credit transaction
dataset = dataset[~dataset['InvoiceNo'].str.contains('C')]

dataset.head()
```

```python
#!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset.columns
```

```python
# separating transactions for country wise - Germany
mybasket = (dataset[dataset['Country']=='Germany'].groupby(['InvoiceNo',
                                                            'Description'])
            ['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
mybasket
```

```python
# converting all positive values to 1 and everything else to be 0

def my_encode_units(x):
    if x <=0:
        return 0
    if x>0:
        return 1
my_basket_sets = mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
my_basket_sets
```

```python
my_freq_item = apriori(my_basket_sets, min_support=0.07, use_colnames=True)
```

```python
my_rules = association_rules(my_freq_item, metric ='lift', min_threshold=2)
```

```python
my_rules.head()
```

```python
my_rules[(my_rules['lift'] >=2) & (my_rules['confidence'] >=0.2)]
```
